### 커스텀 데이터에 맞는 YAML 파일 만들기

```python
import yaml

data = {'train':'/content/length/train/images',
        'val' : '/content/length/valid/images',
        #'names':['없음', '민소매', '반팔', '캡', '7부소매', '긴팔'] ,
        'names' : ['Sleeveless shirt', 'Short sleeve', 'Cap', 'Three-quarter sleeve', 'Long sleeve'],
        'fl_gamma':1.5,
        'nc' : 5
        }
with open('/content/length/length_Data.yaml', 'w') as f:
  yaml.dump(data, f)

with open('/content/length/length_Data.yaml', 'r') as f:
  length_Data = yaml.safe_load(f)
  display(length_Data)
```
